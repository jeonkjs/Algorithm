```python
N, K = list(map(int, input().split()))
```

```python
jew_lst, bag_lst = [], []
for _ in range(N):
    jew_lst.append(list(map(int, input().split())))
    
for _ in range(K):
    bag_lst.append(int(input()))
```

```python
jew_lst.sort()
bag_lst.sort()
```

```python
answer, temp_count = 0, 0
for max_weight in bag_lst:
    for index in range(len(jew_lst)):
        jew_weight, price = jew_lst[index][0], jew_lst[index][1] 
        if max_weight < jew_weight:
            break
        else:
            if temp_count < price:
                temp_count = price
                del_index = index
    answer += temp_count
    temp_count = 0
    del jew_lst[del_index]
```
