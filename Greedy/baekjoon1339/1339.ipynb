```python
num = int(input())

lst = []
for _ in range(num):
    lst.append(input()[::-1])
```

```python
dic = {}
for sentence in lst:
    for index in range(len(sentence)):
        if sentence[index] in dic:
            dic[sentence[index]] = dic[sentence[index]] + (10 ** (index))
        else:
            dic[sentence[index]] = (10**(index))
```

```python
count = 0
num_count = 9
answer = 0
while True:
    if len(dic) == 0:
        break
        
    for word in dic:
        if dic[word] > count:
            count = dic[word]
            large = word
            
    answer += count * num_count
    num_count -= 1
    count = 0
    del dic[large]
```
