```python
num = int(input())

count_1 = 0

nn_lst, pn_lst = [], []
for _ in range(num):
    temp_num = int(input())
    
    if temp_num <= 0:
        nn_lst.append(temp_num)
    elif temp_num > 1:
        pn_lst.append(temp_num)
    elif temp_num == 1:
        count_1 += 1
```

```python
nn_lst.sort()
pn_lst.sort()

print(nn_lst)
print(pn_lst)
```

```python
answer = 0

if len(pn_lst)%2 == 0:
    for index in range(0, len(pn_lst), 2):
        answer += pn_lst[index] * pn_lst[index+1]
elif len(pn_lst)%2 != 0:
    for index in range(1, len(pn_lst), 2):
        answer += pn_lst[index] * pn_lst[index+1]
    answer += pn_lst[0]
     
        
        
if len(nn_lst)%2 == 0:
    for index in range(0, len(nn_lst), 2):
        answer += nn_lst[index] * nn_lst[index+1]
elif len(nn_lst)%2 != 0:
    for index in range(0, len(nn_lst)-1, 2):
        answer += nn_lst[index] * nn_lst[index+1]
    answer += nn_lst[-1]
```

```python
print(answer + count_1)
```
